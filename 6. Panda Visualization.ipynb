```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8-deep')
```

```python
df = pd.read_csv(r"E:\Videos\Ice Cream Ratings.csv")
```

```python
df
```

```python
df = df.set_index('Date')
```

```python
df
```

```python
#Line plot
df.plot() #Default is a line plot
```

```python
df.plot(kind = 'line')
```

```python
#if we want to breat these out by individual column
df.plot(kind = 'line', subplots = True)
```

```python
#Add a title
df.plot(kind = 'line', title = 'Ice Cream Ratings')
```

```python
#Change axis name
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
#Bar plot
df.plot(kind = 'bar')
```

```python
#Stacked bar chart
df.plot(kind = 'bar', stacked = True)
```

```python
#Specify the column
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
#By default, always vertical bar chart, but for horizontal bar chart
df.plot.barh( stacked = True)
```

```python
#Scatter plot
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

```python
#Change the size (s) and colour (c) of the dots in the scatterplot
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'Red')
```

```python
#Histograms(we can specify the bins)
df.plot.hist() #Defaulthistograms bins
```

```python
df.plot.hist(bins = 20) #bins are how many tick marks are on the x-axis
```

```python
#box plot
df.boxplot()
```

```python
#Area plot
df.plot.area()
```

```python
df.plot.area(figsize = (10, 5))
```

```python
#pie chart
df.plot.pie(y = 'Flavor Rating')
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10, 6))
```
