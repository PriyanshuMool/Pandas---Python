```python
import pandas as pd
```

```python
df = pd.read_excel(r"E:\Videos\Customer Call List.xlsx")
```

```python
#notice there are some /, ..., NAN(null values) in the Last_Name column
#There are also different formats of phone numbers
#There are also variation in the Paying Customer column
#Aslo non standardization in the Do_Not_Contact
#We need to standardize each column respectively
#We want to also get rid of Not_Useful_Column
```

```python
#Step 1: Drop the duplicates from the dataset
df.drop_duplicates()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
#Step 2: Remove the columns we do not need (here, Not_Useful_Column)
df.drop(columns = 'Not_Useful_Column')
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
#Step 3: There are obvious variations and null values in the Last_Name column, we should clean up them
#Use strip
df['Last_Name'].str.lstrip('...') #lstrip() strips off the specified values from the left side of the string
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('...') #this saves the stripping
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.rstrip('_') #rstrip() strips off the specified values from the right side of the string
```

```python
df
```

```python
#The other fast methof would have been: df['Last_Name'] = df['Last_Name'].str.strip("123._/")
```

```python
#Step 4: Working with phone numbers
df
#We need all the phone numbers to look in this formate '123-545-5421'
#We will get rid of all the non numeric value, strip dowm to just numbers(no / or - or anything)
```

```python
df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True) #We are taking all of alphabets(a-z, A-Z) and numbers(0-9) and replacing it with nothing('')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
```

```python
df
```

```python
#Step 5: To format the Phone_Number column by using the lambda
df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #We want first 3 strings of phone number the add a slash or a dash then next 3 and add a slash or a dash
```

```python
#The values in Phone_Number is float type but not string type, so we need to first convert them to string type
df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
#Replace the nan-- values
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '') #replacing 'nan--' with nothing
```

```python
df
```

```python
#Doing the same for Na--
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
```

```python
df
```

```python
#Splitting columns: address column needs to be splited into 3 columns
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2,  expand=True) #1.splitting on comma 2.next specify the values from left to right it should look for
```

```python
df #added to the end of the dataframe
```

```python
#Step 6: Replace all the Paying Customer values to all Y and N (nothing else)
df['Paying Customer'].str.replace('Yes', 'Y')
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
df
```

```python
#Step 7: Doing the same for Do_Not_Contact column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df
```

```python
df = df.replace('N/a', '')
df = df.replace('NaN', '') #technically NaN is empty is its going to nothing to those values
```

```python
df
```

```python
#But we can fill those NaN values with blank
df = df.fillna('')
```

```python
df
```

```python
#remove the phone numbers of the people who have responded Y to Do_Not_Contact column
#loop through the Do_Not_Contact based of index
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
```

```python
df
```

```python
#do the exact same thing for phone number, if blank then don't contact
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
```

```python
df
```

```python
#we should reset the index
df.reset_index(drop = True) #drops the original index and replaces with new index
```

```python
df = df.reset_index(drop = True)
```
