```python
#Group By and Aggregate Functions in Pandas
```

```python
#Gorup by: going to group together the values within a column and display them all on a same row
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r"E:\Videos\Flavors.csv")
```

```python
df
```

```python
#Let's group by the Base Flavor
df.groupby('Base Flavor')
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only = True) #took the base flavors (chocolate and Vanilla) and too the mean(avreage) of all the columns that have integers
```

```python
df.groupby('Base Flavor').mean(numeric_only = True)
```

```python
df.groupby('Base Flavor').count()
```

min() and max() functions take the strings as well. They look at the string as A/a with least and Z/z with max values.

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').sum(numeric_only = True)
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'min', 'sum'], 'Texture Rating': ['mean', 'max', 'min', 'sum']}) 
```

```python
#groupby on multiple columns
#Let's group base flavor and liked columns
df.groupby(['Base Flavor', 'Liked']).mean(numeric_only = True)
```

```python
df.groupby('Base Flavor').describe()
```
